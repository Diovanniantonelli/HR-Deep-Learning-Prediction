Pré-Processamento de Dados

```python
employee_df.head()
```

```python
X_cat2 = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat2
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
X_cat2 = onehotencoder.fit_transform(X_cat2).toarray()
```

```python
X_cat2.shape
```

```python
X_cat2 = pd.DataFrame(X_cat2)
type(X_cat2)
```

```python
X_cat2
```

```python
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'TotalWorkingYears','EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'OverTime','PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
X_numerical
```

```python
X_all = pd.concat([X_cat2, X_numerical], axis=1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler

# Convert the column names of X_cat2 to strings
X_cat2.columns = X_cat2.columns.astype(str)

# Concatenate the dataframes
X_all = pd.concat([X_cat2, X_numerical], axis=1)

# Now fit_transform should work without the TypeError
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
X
```

```python
y = employee_df['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
#
```

REGRESSÃO LOGISTICA

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
#Precision
15 / (15+7)
```

```python
#Recall
15 / (15+48)
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score


precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='macro')
```

```python
print(classification_report(y_test, y_pred))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
y_pred
```

```python
y_test
```

Redes Neurais Artificiais

```python
X_train.shape
```

```python
(50+1)/2
```

```python
import tensorflow as tf
```

```python
rede_neural = tf.keras.models.Sequential()
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu', input_shape=(50,))) #Dense neuronios interligados
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
rede_neural.fit(X_train, y_train, epochs=200)
```

```python
y_pred = rede_neural.predict(X_test)
y_pred
```

```python
y_pred = (y_pred >= 0.5)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True);
```

```python
print(classification_report(y_test, y_pred))
```
